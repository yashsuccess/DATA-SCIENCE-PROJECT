# AIM - AIM of this Project is to Scrap the Two Wheeler Data from Flipkart Website

# STEP-1 Importing required Libaries 

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# STEP-2 Accessing The Flipkart Website

```python
url="https://www.flipkart.com/two-wheelers-store?fm=neo%2Fmerchandising&iid=M_fa8fd3b8-ae7a-42be-890a-ec3225a835bf_1_372UD5BXDFYS_MC.SCJPZ04TFJJM&otracker=hp_rich_navigation_1_1.navigationCard.RICH_NAVIGATION_Two%2BWheelers~Petrol%2BVehicles_SCJPZ04TFJJM&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_1_L1_view-all&cid=SCJPZ04TFJJM"
print(url)
```

```python
r=requests.get(url)
htmlcontent=r.content
print(htmlcontent)
```

# STEP-3 Beautifying the HTML Code

```python
soup=BeautifulSoup(htmlcontent,"html.parser")
print(soup)
```

```python
soup.prettify
```

# i) Accessing the Product Name

```python
product_name=soup.find_all(class_="s1Q9rs")
print(product_name)
```

```python
product_names=[]
for i in range(0,len(product_name)):
    product_names.append(product_name[i].get_text())
    
print(product_names)
len(product_names)
```

# ii) Accessing the product Price

```python
product_price=soup.find_all(class_="_30jeq3")
print(product_price)
```

```python
product_prices=[]
for i in range(0,len(product_price)):
    if i<=17:
        product_prices.append(product_price[i].get_text())
    
print(product_prices)
len(product_prices)
```

# iii)Accessing the Product Review Rating

```python
product_Review_Rating=soup.find_all(class_="_3LWZlK")
print(product_Review_Rating)
```

```python
product_Review_Ratings=[]
for i in range(0,len(product_Review_Rating)):
    if i<=17:
        product_Review_Ratings.append(product_Review_Rating[i].get_text())
    
print(product_Review_Ratings)
len(product_Review_Ratings)
```

# STEP-4 Converting the data into DataFrame

```python
df=pd.DataFrame({"Product_name":product_names,"Product_Price":product_prices,"Product_Review_Rating":product_Review_Ratings})
df
```

# STEP-5 Storing all the Data in CSV File and Opening the file in the application

```python
df.to_csv("Web Scraping Project.csv",index=False)
a=pd.read_csv("Web Scraping Project.csv")
a
```
